# Group by and aggregation 

Grouper un set de données suivant une ou plusieurs dimensions et appliquer une réduction est une composante clé de toute analyse. 

Le langage SQL ainsi que les tableaux croisés dynamiques offrent de premières fonctions de Grouping. 

Dans ce chapitre nous verrons comment Pandas permet de réaliser de séparer un set de données suivant une ou plusieurs clés et appliqué des réductions ou transformations sur chacun de ces groupes. 

```python
import pandas as pd
```

```python
reserve = pd.read_csv("data/reserve_parlementaire.csv")

reserve.sort("Subvention allouée", ascending=False)
```

```python
sub_parlementaire = reserve.groupby("Parlementaire transmetteur").sum()["Subvention allouée"].order(ascending=False)


sub_parlementaire.head()
```

```python
reserve.groupby("Bénéficiaire").size().order(ascending=False)[0:10]
```

```python
print reserve.groupby("Groupe politique du parlementaire").sum()["Subvention allouée"].order(ascending=False)
```
